```python
# import Dependencies
import pandas as pd
import numpy as np
```

```python
# bring back in limted NYC PLUTO data
csv_path = "nyc_pluto_limit.csv"
limit_nyc_pluto = pd.read_csv(csv_path, encoding="utf-8")
```

```python
limit_nyc_pluto.head()
```

```python
# use pd.dummies to break down 'land use' into separarte columns
limit_nyc_pluto = pd.get_dummies(limit_nyc_pluto, columns=['landuse'])
```

```python
limit_nyc_pluto.head()
```

```python
# multiply bldgarea by value in landuse column
limit_nyc_pluto['landuse_1.0'] = limit_nyc_pluto['landuse_1.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_2.0'] = limit_nyc_pluto['landuse_2.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_3.0'] = limit_nyc_pluto['landuse_3.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_4.0'] = limit_nyc_pluto['landuse_4.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_5.0'] = limit_nyc_pluto['landuse_5.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_6.0'] = limit_nyc_pluto['landuse_6.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_7.0'] = limit_nyc_pluto['landuse_7.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_8.0'] = limit_nyc_pluto['landuse_8.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_9.0'] = limit_nyc_pluto['landuse_9.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_10.0'] = limit_nyc_pluto['landuse_10.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto['landuse_11.0'] = limit_nyc_pluto['landuse_11.0'].multiply(limit_nyc_pluto['bldgarea'], axis="index")
limit_nyc_pluto.head()
```

```python
list(limit_nyc_pluto.columns.values)
```

```python
# create limited dataframe for only landuse
nyc_pluto_landuse = limit_nyc_pluto[['borough', 'zipcode', 'landuse_1.0', 'landuse_2.0','landuse_3.0', 'landuse_4.0', 'landuse_5.0', 'landuse_6.0', 'landuse_7.0', 'landuse_8.0', 'landuse_9.0', 'landuse_10.0', 'landuse_11.0']].copy()
nyc_pluto_landuse.head()
```

```python
nyc_pluto_landuse = nyc_pluto_landuse.groupby(['zipcode', 'borough']).sum()
nyc_pluto_landuse.head()
```

```python
# add a total column - for all land use per zip code
nyc_pluto_landuse['total_bldgarea'] = nyc_pluto_landuse['landuse_1.0'] + nyc_pluto_landuse['landuse_2.0'] + nyc_pluto_landuse['landuse_3.0'] + nyc_pluto_landuse['landuse_4.0'] + nyc_pluto_landuse['landuse_5.0'] + nyc_pluto_landuse['landuse_6.0'] + nyc_pluto_landuse['landuse_7.0'] + nyc_pluto_landuse['landuse_8.0'] + nyc_pluto_landuse['landuse_9.0'] + nyc_pluto_landuse['landuse_10.0'] + nyc_pluto_landuse['landuse_11.0']
nyc_pluto_landuse.head()
```

```python
# save as csv
nyc_pluto_landuse.to_csv('nyc_pluto_landuse.csv')
```
