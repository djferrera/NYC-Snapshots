```python
# limited data
csv_path_2 = "/Users/djferrera/Documents/GitHub/Columbia Boot Camp/Github/NYC Snapshots/data/pluto_19v2_limit.csv"
nyc_pluto_limit = pd.read_csv(csv_path_2, encoding="utf-8")
```

```python
# view data - limited, more managable
nyc_pluto_limit.head()
```

```python
# new column for neighborhood - based on zip code
# nyc_pluto_limit['neighborhood'] = nyc_pluto_limit['zipcode']
# nyc_pluto_limit.head()
```

```python
# new column for landuse
nyc_pluto_limit['use'] = nyc_pluto_limit['landuse']
```

```python
nyc_pluto_limit['use'].replace(1.0, 'Central One & Two Family Buildings',inplace=True)
nyc_pluto_limit['use'].replace(2.0, 'Multi-Family Walk-Up Buildings',inplace=True)
nyc_pluto_limit['use'].replace(3.0, 'Multi-Family Elevator Buildings',inplace=True)
nyc_pluto_limit['use'].replace(4.0, 'Mixed Residential & Commercial Buildings',inplace=True)
nyc_pluto_limit['use'].replace(5.0, 'Commercial & Office Buildings',inplace=True)
nyc_pluto_limit['use'].replace(6.0, 'Industrial & Manufacturing',inplace=True)
nyc_pluto_limit['use'].replace(7.0, 'Transportation & Utility',inplace=True)
nyc_pluto_limit['use'].replace(8.0, 'Public Facilities & Institutions',inplace=True)
nyc_pluto_limit['use'].replace(9.0, 'Open Space & Outdoor Recreation',inplace=True)
nyc_pluto_limit['use'].replace(10.0, 'Parking Facilities',inplace=True)
nyc_pluto_limit['use'].replace(11.0, 'Vacant Land',inplace=True)
nyc_pluto_limit.head()
```

```python
# save as csv
nyc_pluto_limit.to_csv('nyc_pluto_limit.csv')
```
