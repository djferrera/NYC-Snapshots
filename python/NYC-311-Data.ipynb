```python
# NYC 311 Data
# 311 Service Requests from 2010 to Present
# https://data.cityofnewyork.us/Social-Services/311-Service-Requests-from-2010-to-Present/erm2-nwe9
```

```python
# import Dependencies
import pandas as pd
import numpy as np
```

```python
# data
csv_path = "311_Service_Requests_from_2010_to_Present.csv"
nyc_311 = pd.read_csv(csv_path, encoding="utf-8")
```

```python
# remove time from date column
nyc_311['Created Date'] = nyc_311['Created Date'].str[:10]
nyc_311.head()
```

```python
# actually, look at only year

nyc_311['Year'] = nyc_311['Created Date'].str[-4:]
nyc_311.head()
```

```python
# narrow data down to only 2017
nyc_311_2017 = nyc_311.loc[nyc_311['Year'] == '2017']
nyc_311_2017.head()
```

```python
# check column names
nyc_311_2017.columns
```

```python
# narrow down data even further
nyc_311_2017_limit = nyc_311_2017.drop(['Unique Key', 'Closed Date', 'Agency Name', 'Location Type', 'Incident Address', 'Street Name', 'Cross Street 1', 'Cross Street 2', 'Intersection Street 1', 'Intersection Street 2', 'City','Landmark', 'Status', 'Due Date', 'Resolution Description', 'Resolution Action Updated Date', 'Community Board', 'BBL', 'X Coordinate (State Plane)', 'Y Coordinate (State Plane)', 'Open Data Channel Type', 'Park Facility Name', 'Taxi Company Borough', 'Taxi Pick Up Location', 'Bridge Highway Name', 'Road Ramp', 'Bridge Highway Direction', 'Latitude', 'Longitude', 'Location', 'Year'], axis=1)
```

```python
# remove a few more columns
nyc_311_2017_limit = nyc_311_2017_limit.drop(['Address Type', 'Facility Type', 'Park Borough', 'Vehicle Type', 'Bridge Highway Segment'], axis=1)
```

```python
# preview data
nyc_311_2017_limit.head()
```

```python
# change 'incident zip' to 'zipcode' (for merge)
nyc_311_2017_limit = nyc_311_2017_limit.rename(columns={'Incident Zip': 'zipcode'})
nyc_311_2017_limit.head()
```

```python
# bring in PLUTO data to merge on zip code
# narrow out the non-city zip codes
# bring back in limted NYC PLUTO data
csv_path2 = "nyc_pluto_landuse.csv"
nyc_pluto_landuse = pd.read_csv(csv_path2, encoding="utf-8")
```

```python
nyc_pluto_landuse.head()
```

```python
# merge
nyc_311_2017_limit = pd.merge(nyc_311_2017_limit, nyc_pluto_landuse, on='zipcode', how='inner')
nyc_311_2017_limit.head()
```

```python
# check column names
nyc_311_2017_limit.columns
```

```python
# remove excess columns
nyc_311_2017_limit = nyc_311_2017_limit.drop(['Borough', 'landuse_1.0', 'landuse_2.0','landuse_3.0','landuse_4.0','landuse_5.0','landuse_6.0','landuse_7.0','landuse_8.0','landuse_9.0','landuse_10.0','landuse_11.0','total_bldgarea'], axis=1)
```

```python
# preview data
nyc_311_2017_limit.head()
```

```python
# number of rows
nyc_311_2017_limit.shape
```

```python
# unique values in 'Complaint Type'
nyc_311_2017_limit['Complaint Type'].nunique()
```

```python
# remove values that appear only once
nyc_311_2017_min = nyc_311_2017_limit[nyc_311_2017_limit['Complaint Type'].isin(nyc_311_2017_limit['Complaint Type'].value_counts()[nyc_311_2017_limit['Complaint Type'].value_counts()>500].index)]
nyc_311_2017_min.head()
```

```python
# number of rows
nyc_311_2017_min.shape
```

```python
# unique values in 'Complaint Type'
nyc_311_2017_min['Complaint Type'].nunique()
```

```python
# save as csv
nyc_311_2017_min.to_csv('nyc_311_2017_min.csv')
```
